```python
import cv2 as cv
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img=cv.imread("tree.jpg")
kernal=np.ones((3,3),np.float32)/9
filter=cv.filter2D(img,-1,kernal)
cv.imshow("image",img)
cv.imshow("filter2d",filter)
cv.waitKey(0)
cv.destroyAllWindows()
```

# bluring

```python
img=cv.imread("tree.jpg")
blur=cv.blur(img,(5,5))
cv.imshow("image",img)
cv.imshow("blur",blur)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Gaussian Blur

```python
img=cv.imread("noise.jpeg")
Gblur=cv.GaussianBlur(img,(5,5),0)
cv.imshow("image",img)
cv.imshow("Gaussianblur",Gblur)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Bilateral Filter

```python
img=cv.imread("noise.jpeg")
bFilter=cv.bilateralFilter(img,10,20,20)
cv.imshow("image",img)
cv.imshow("bilateralFilter",bFilter)
cv.waitKey(0)
cv.destroyAllWindows()
```

# median Filters

```python
img=cv.imread("salt1.jpg")
m=cv.medianBlur(img,5)
cv.imshow("image",img)
cv.imshow("mideanFilter",m)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Laplacian Filter

```python
img=cv.imread("noise.jpeg",0)
lap=cv.Laplacian(img,cv.CV_64F,ksize=3)
cv.imshow('image',img)
cv.imshow('laplacian',lap)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Sobel Filter

```python
img=cv.imread("noise.jpeg",0)
sobelx=cv.Sobel(img,cv.CV_32F,1,0,ksize=3)
print(sobelx)
cv.imshow('image',img)
cv.imshow("sobelx",sobelx)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img=cv.imread("noise.jpeg",0)
sobely=cv.Sobel(img,cv.CV_32F,0,1,ksize=3)
print(sobely)
cv.imshow('image',img)
cv.imshow("sobely",sobely)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img=cv.imread("noise.jpeg",0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
lap=cv.Laplacian(img,cv.CV_32F,ksize=3)
lap=np.uint8(np.absolute(lap))
sobelx=cv.Sobel(img,cv.CV_32F,1,0,ksize=3)
sobely=cv.Sobel(img,cv.CV_32F,0,1,ksize=3)
sobelx=np.uint8(np.absolute(sobelx))
sobely=np.uint8(np.absolute(sobely))
```

```python
titles=['image','laplacian','soblex','sobely']
images=[img,lap,sobelx,sobely]
for i in range(len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
plt.show
```
